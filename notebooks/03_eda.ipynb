#### Histograma + densidad

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns
numeric_columns = numeric_columns.drop("ID")
```

```python
numeric_columns
```

```python
for column in numeric_columns:
    plt.figure(figsize=(10,6))
    sns.histplot(data=df, x=column, kde=True, stat="density", color="green")
    plt.title(f"Histograma de la variable: {column}")
    plt.show()
```

```python
for column in numeric_columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data=df, x=column, color="salmon")
    plt.title(f"Boxplot de la variable: {column}")
    plt.show()
```

```python
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    col_mean = df[col].mean()
    df.loc[(df[col] < lower) | (df[col] > upper), col] = col_mean
```

```python
for column in numeric_columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data=df, x=column, color="salmon")
    plt.title(f"Boxplot de la variable: {column}")
    plt.show()
```

```python
for column in numeric_columns:
    plt.figure(figsize=(10,6))
    sns.histplot(data=df, x=column, kde=True, hue="target", stat="density", color="green")
    plt.title(f"Histograma de la variable: {column}")
    plt.show()
```

```python
corr_matrix = df[numeric_columns].corr(method="pearson")
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix, 
    annot=True,          
    fmt=".2f",           
    cmap="coolwarm",     
    center=0,            
    linewidths=0.5,      
    cbar_kws={"shrink": .8}  
)

plt.title("Matriz de Correlación (Pearson)", fontsize=14, pad=15)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 1. Distribucion Normal

```python
from scipy import stats
```

```python
mu, sigma = 0, 0.5
```

```python
normal = stats.norm(mu, sigma)
```

```python
x = np.linspace(normal.ppf(0.01), normal.ppf(0.99),100000)
```

```python
fp = normal.pdf(x)
```

```python
plt.plot(x,fp)
```

##### 2. Distribucion Uniforme

```python
uniforme = stats.uniform()
```

```python
x = np.linspace(uniforme.ppf(0.01), uniforme.ppf(0.99),100000)
```

```python
fp = uniforme.pdf(x)
```

```python
plt.plot(x,fp)
```
