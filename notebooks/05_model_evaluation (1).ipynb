```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/tinch/coderhouse/template_model/data/processed/dataset_merged_processed_no_outliers_numeric.csv")
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head(3)
```

```python
X = df.drop(columns=["target","ID"])
y = df["target"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

#### arbol de decision

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier(max_depth=3, random_state=42)
```

```python
%%time
decision_tree.fit(X_train, y_train) # siempre que usemos .fit() en algun objeto, USAMOS TRAIN.
```

```python
%%time
y_pred_train_dt = decision_tree.predict(X_train) # devuelve 0 y 1, por defecto cuando la prob > 0.5
y_pred_test_dt = decision_tree.predict(X_test)
```

```python
y_pred_train_dt
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("train")
print(classification_report(y_train,y_pred_train_dt))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test_dt))
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier()
```

```python
%%time
knn.fit(X_train_scaled, y_train)
```

```python
%%time
y_pred_train_knn = knn.predict(X_train_scaled)
y_pred_test_knn = knn.predict(X_test_scaled)
```

```python
print("train")
print(classification_report(y_train,y_pred_train_knn))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test_knn))
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=200, max_depth=3, random_state=42)
```

#### xgboost

```python
from xgboost import XGBClassifier
```

```python
xgboost = XGBClassifier(max_depth=3, eval_metric="logloss", random_state=42)
```

```python
%%time
xgboost.fit(X_train, y_train)
```

```python
%%time
y_pred_train_xgb = xgboost.predict(X_train) 
y_pred_test_xgb = xgboost.predict(X_test)
```

```python
print("train")
print(classification_report(y_train,y_pred_train_xgb))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test_xgb))
```

```python
# optimizacion de xgboost
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": [200, 300],
    "max_depth": [1, 3],
    "learning_rate": [0.01, 0.05]
}
```

```python
xgboost = XGBClassifier(eval_metric="logloss", random_state=42)
```

```python
grid_search = GridSearchCV(estimator=xgboost, param_grid = param_grid, scoring="precision", cv=3, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model
```

```python
y_pred_train = best_model.predict(X_train)
```

```python
y_pred_test = best_model.predict(X_test)
```

```python
print("train")
print(classification_report(y_train,y_pred_train))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
probabilidades_train = best_model.predict_proba(X_train)[:,1]
```

```python
probabilidades_test = best_model.predict_proba(X_test)[:,1]
```

```python
plt.figure(figsize=(10,6))
sns.histplot(probabilidades_train, bins=100,kde=True,label="Train", stat="density", alpha=0.7)
sns.histplot(probabilidades_test, bins=100,kde=True,label="Test", stat="density", alpha=0.7)
plt.axvline(0.5, color="red", linestyle="--")
plt.show()
```

```python
threshold = 0.63
```

```python
y_train_pred_custom = (probabilidades_train >= threshold).astype(int)
```

```python
y_test_pred_custom = (probabilidades_test >= threshold).astype(int)
```

```python
print("train")
print(classification_report(y_train,y_train_pred_custom))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_test_pred_custom))
```
