```python
import pandas as pd
import numpy as np
```

```python
ruta = "C:/Users/tinch/coderhouse/template_model/data/raw/"
```

```python
cardio = pd.read_excel(f"{ruta}cardio.xlsx")
```

```python
fisicas = pd.read_excel(f"{ruta}fisicas.xlsx")
```

```python
metabolicas = pd.read_excel(f"{ruta}metabolicas.xlsx")
```

```python
oral = pd.read_excel(f"{ruta}oral.xlsx")
```

```python
sentidos = pd.read_excel(f"{ruta}sentidos.xlsx")
```

```python
target = pd.read_excel(f"{ruta}target.xlsx")
```

```python
df = pd.merge(cardio, fisicas, on ="ID", how="left")
```

```python
df = pd.merge(df, metabolicas, on ="ID", how="left")
```

```python
df = pd.merge(df, oral, on ="ID", how="left")
```

```python
df = pd.merge(df, sentidos, on ="ID", how="left")
```

```python
df = pd.merge(df, target, on ="ID", how="left")
```
